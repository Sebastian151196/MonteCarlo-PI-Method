## Esta es una aproximación númerica de Pi, basado en el método de Monte Carlo usando un circulo unitario inscrito dentro de un polígono. 

```python
import numpy as np
import matplotlib.pyplot as plt
import time as time
%matplotlib inline 

total_random_points = int(input("Ingrese la cantidad de puntos para estimar Pi: "))
inside_circle = 0.0

start_time = time.time()
x_plot_array = np.empty(shape=(1, total_random_points))
y_plot_array = np.empty(shape=x_plot_array.shape)

#Crearemos un loop para recorrer todos los puntos
for i in range(total_random_points):
    x = np.random.rand()
    x_plot_array = np.append(x_plot_array, [x])
    y = np.random.rand()
    y_plot_array = np.append(y_plot_array, [y])
    x_squared = x**2
    y_squared = y**2
    if np.sqrt(x_squared + y_squared) < 1.0:
        inside_circle += 1


pi_approx = (inside_circle / total_random_points) * 4
print("Valor estimado por MC: %.16f" % pi_approx)
print("La diferencia entre el número estimado y el número exacto Pi: %.16f" % (abs(pi_approx-np.pi)))
print("Tiempo de Finalización: %f segundos" % (time.time() - start_time))

random_points_plot = plt.scatter(x_plot_array, y_plot_array, color='gray', s=.5) # Puntos
circle_plot = plt.Circle((0, 0), 1, color='red', linewidth=2, fill=False)

ax = plt.gca()
ax.cla()
ax.set_aspect('equal')

plt.title('Numero de pruebas N= %d ' ' Estimate of PI = %.16f' % (total_random_points, pi_approx))
plt.xlim(-1.0, 1.0)
plt.ylim(-1.0, 1.0)

ax.add_artist(random_points_plot)
ax.add_artist(circle_plot)

plt.show()
```
